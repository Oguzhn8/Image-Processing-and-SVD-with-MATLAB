```python
import cv2
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import svd

from numpy import array
from numpy import diag
from numpy import zeros
```

```python
# Read Images
nasacolor = mpimg.imread('TarantulaNebula.jpg')


# Output Images
plt.imshow(nasacolor)
```

```python
nasa_sum = np.sum(nasacolor, axis=2) #sum up red+green+blue
max_num=np.max(np.max(nasa_sum))     #find the max value
nasa_new=nasa_sum*255/max_num;       #make this be bright white

plt.imshow(nasa_new, cmap=plt.get_cmap("gray"))
```

```python
u, s, vh = svd(nasa_new)
```

```python
plt.semilogy(np.diag(s))
```

```python
u, s, vh = svd(nasa_new)
# create m x n Sigma matrix
Sigma = zeros((nasa_new.shape[0], nasa_new.shape[1]))
# populate Sigma with n x n diagonal matrix
Sigma[:nasa_new.shape[0], :nasa_new.shape[0]] = diag(s)

n_elements = 100 #For versions 25 and 50 please change 100
Sigma = Sigma[:, :n_elements]
vh = vh[:n_elements, :]
# reconstruct
nasa100 = u.dot(Sigma.dot(vh))
plt.title('100 SV')
plt.imshow(nasa100, cmap=plt.get_cmap("gray"))
```
